# Training Loss

```python
import pandas as pd
import matplotlib.pyplot as plt
from gensim.models import KeyedVectors

plt.style.use(['seaborn-whitegrid', 'seaborn-poster'])
```

## Skip-gram Loss

```python
loss_file = './model-skipgram/logs/train-loss-e300-w5-i100-t080120054147.csv'
with open(loss_file, 'r') as fin:
    loss = fin.read()
    
loss = loss.split(',')
loss = list(map(lambda x: float(x), loss))

plt.plot(loss)
plt.title('Skip-gram - Embedding Size 150 - Window Size 5')
plt.xlabel('Epochs')
plt.ylabel('Loss')
```
